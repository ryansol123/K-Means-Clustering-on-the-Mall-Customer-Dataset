## K Means Clustering 

```python
from sklearn.cluster import KMeans 
from sklearn.preprocessing import MinMaxScaler
```

```python
# Scatter plot of spending score based on income 

df.plot(kind='scatter', x='Spending Score', y='Salary (k$)', title='Spending Score v Salary (k$)')
```

```python
km = KMeans(n_clusters=5)     # looking at the scatter plot, there's 5 distinct clusters, hence the selection of 5.

y_predicted = km.fit_predict(df[['Spending Score','Salary (k$)']])  # fit n predict SS and Salary. Used K Means algo on age and income. Computed the cluster based on our assumption of 5.
y_predicted
```

```python
df['cluster']=y_predicted  # visualising array
df.head()
```

```python
km.cluster_centers_
```

```python
# Visualising the cluster centers.

sns.scatterplot(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
sns.scatterplot(df['Spending Score'], df['Salary (k$)'], hue = df['cluster'])
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
```

## Preprocessing

```python
# To scale, and therefore cluster correctly, do preprocessing.

scaler = MinMaxScaler()      # Scales each feature to a given range.

scaler.fit(df[['Spending Score']])    # Gives the scale a range from 0 to 1.
df['Spending Score'] = scaler.transform(df[['Spending Score']])
df   # Put spending score in the range of 0 to 1.

#scaler.fit(df[['Salary (k$)']])
#df['Salary (k$)'] = scaler.transform(df[['Salary (k$)']])
df
```

## Elbow Plot Method

```python
sse = []
k_rng = range(1,11)  # range at which the elbow extends to. 
for k in k_rng:
    km = KMeans(n_clusters=k) # through each iteration, create a new model with clusters k.
    km.fit(df[['Spending Score','Salary (k$)']])   # fitting the 2 variables, Spending Score and Salary.
    sse.append(km.inertia_)  # inertia = Gives you SSE. Appending it to the array sse.
    
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
